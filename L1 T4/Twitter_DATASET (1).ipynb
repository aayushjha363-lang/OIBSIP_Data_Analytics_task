```python


import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns


import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python
data=pd.read_csv("/kaggle/input/twitter-tweets-sentiment-dataset/Tweets.csv")
data.head()
```

```python
data.drop( 'textID', axis=1, inplace=True)
```

```python
fig = plt.figure(figsize=(7,5))
sns.countplot(x="sentiment",data=data)
```

```python
fig = plt.figure(figsize=(7,7))
colors = ("yellowgreen", "gold", "red")
wp = {'linewidth':2, 'edgecolor':"black"}
tags = data['sentiment'].value_counts()
explode = (0.1,0.1,0.1)
tags.plot(kind='pie', autopct='%1.1f%%', shadow=True, colors = colors,
         startangle=90, wedgeprops = wp, explode = explode, label='')
plt.title('Distribution of sentiments')
```

```python
from wordcloud import WordCloud
text = " ".join(i for i in data[data['sentiment']=='positive']['selected_text'])
wordcloud = WordCloud( background_color="white").generate(text)

plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title('wordcloud for positive review')
plt.show()
```
